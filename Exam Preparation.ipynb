## Question 1 — Call Option (Binomial Model)

Consider a **2-period binomial model** for a stock currently trading at **$48**.  In each period, the stock price can either move **up by 25%** or **down by 25%**. The **risk-free rate is 6% per period**.

a) Calculate the **risk-neutral probabilities** $(p \text{ and } 1-p)$.

b) Calculate the **fair value of a European Call Option** with a **strike price of $50** expiring in **2 periods**.

c) What is the **fair value of an American Call Option** with the same characteristics at $t = 0$?

---


## Solution 1

### a) Risk-neutral probabilities
$$
p = \frac{(1+r)-d}{u-d}
= \frac{1.06 - 0.75}{1.25 - 0.75}
= \frac{0.31}{0.50}
= 0.62
$$

$$
1 - p = 0.38
$$


### b) European Call Option Value \((K = 50)\)

#### 1. Stock price tree
$$
S_0 = 48
$$

$$
S_u = 60, \quad S_d = 36
$$

$$
S_{uu} = 75, \quad S_{ud} = 45, \quad S_{dd} = 27
$$

#### 2. Call payoffs at maturity $(t=2)$
$$
C_{uu} = \max(75 - 50, 0) = 25
$$

$$
C_{ud} = \max(45 - 50, 0) = 0
$$

$$
C_{dd} = \max(27 - 50, 0) = 0
$$

#### 3. Backward induction t=1

At node $u, S=60 $:
$$
C_u = \frac{0.62(25) + 0.38(0)}{1.06}
= \frac{15.5}{1.06}
= 14.6226
$$
At node $d, S=36$:
$$
C_d = 0
$$

#### 4. Value at t=0
$$
C_0 = \frac{0.62(14.6226) + 0.38(0)}{1.06}
= \frac{9.0660}{1.06}
= 8.5529
$$
**Fair Value (European Call): $\boxed{8.55}$**



### c) American Call Option Value

Check early exercise at t=1:

- At node $u$:  
  Intrinsic value $= 60 - 50 = 10$ 
  Continuation value $= 14.6226$ → **Do not exercise**

- At node $d$:  
  Intrinsic value $= 0$  
  Continuation value $= 0$ → **Indifferent**

Early exercise is **not optimal**, so:

$$
C_0^{(A)} = C_0^{(E)} = \boxed{8.55}
$$

**Fair Value (American Call): $\boxed{8.55}$**

## Question 2 — 2-Period Binomial Option Pricing

A stock is currently trading at **$55**. Over each of the next **two periods**, the stock price can either:

- move **up** by a factor of $u = 1.30$, or  
- move **down** by a factor of $d = 0.85$.

The **risk-free rate is 4% per period** $(r = 0.04)$.

You are asked to value options with **strike price $K = 58$ and maturity $T=2$ periods**.

a) Draw the **2-period stock price tree** and compute the terminal stock prices $S_{uu}, S_{ud}, S_{dd}$.

b) Compute the **risk-neutral probability** $p$.

c) Compute the **fair value at $t=0$** of:
   - (i) a **European call**, and  
   - (ii) a **European put**.

d) Verify **put–call parity** at $t=0$:
$$
C_0 - P_0 \stackrel{?}{=} S_0 - \frac{K}{(1+r)^2}.
$$

---

## Solution 2

#### Given
$$
S_0 = 55,\quad u=1.30,\quad d=0.85,\quad r=0.04,\quad K=58,\quad T=2,\quad (1+r)=1.04
$$

### a) Stock price tree (2 periods)

Time 1:
$$
S_u = 55(1.30)=71.50,\quad S_d = 55(0.85)=46.75
$$

Time 2:
$$
S_{uu}=55(1.30)^2=92.95
$$

$$
S_{ud}=55(1.30)(0.85)=60.775
$$

$$
S_{dd}=55(0.85)^2=39.7375
$$


### b) Risk-neutral probability

$$
p=\frac{(1+r)-d}{u-d}
=\frac{1.04-0.85}{1.30-0.85}
=\frac{0.19}{0.45}
=0.4222
$$

$$
1-p=0.5778
$$


### c) European option values

### Terminal payoffs at $t=2$

**Call** payoff $C=\max(S-K,0)$:
$$
C_{uu}=\max(92.95-58,0)=34.95
$$

$$
C_{ud}=\max(60.775-58,0)=2.775
$$

$$
C_{dd}=\max(39.7375-58,0)=0
$$

**Put** payoff $P=\max(K-S,0)$:
$$
P_{uu}=0,\quad P_{ud}=0,\quad P_{dd}=58-39.7375=18.2625
$$


### Backward induction to $t=1$

Discount factor per period: divide by $1.04$.

**Call at time 1**
$$
C_u=\frac{pC_{uu}+(1-p)C_{ud}}{1.04}
=\frac{0.4222(34.95)+0.5778(2.775)}{1.04}
=15.7308
$$

$$
C_d=\frac{pC_{ud}+(1-p)C_{dd}}{1.04}
=\frac{0.4222(2.775)+0.5778(0)}{1.04}
=1.1266
$$

**Put at time 1**
$$
P_u=\frac{pP_{uu}+(1-p)P_{ud}}{1.04}=0
$$

$$
P_d=\frac{pP_{ud}+(1-p)P_{dd}}{1.04}
=\frac{0.5778(18.2625)}{1.04}
=10.1458
$$


### Value at $t=0$

**European Call**
$$
C_0=\frac{pC_u+(1-p)C_d}{1.04}
=\frac{0.4222(15.7308)+0.5778(1.1266)}{1.04}
=7.0123
$$

**European Put**
$$
P_0=\frac{pP_u+(1-p)P_d}{1.04}
=\frac{0.4222(0)+0.5778(10.1458)}{1.04}
=5.6366
$$

$$
\boxed{C_0 \approx 7.01}\qquad \boxed{P_0 \approx 5.64}
$$


### d) Put–call parity check

$$
C_0 - P_0 = 7.0123 - 5.6366 = 1.3757
$$

$$
S_0 - \frac{K}{(1.04)^2}
= 55 - \frac{58}{1.0816}
= 55 - 53.6243
= 1.3757
$$

Parity holds (up to rounding):
$$
\boxed{C_0 - P_0 = S_0 - \frac{K}{(1+r)^2}}
$$

## Question 3 – Two Stocks with Single-Index Regressions

An analyst estimates the following single-index regressions of excess returns on the market’s excess returns:

| Stock | Alpha $\alpha$ | Beta $\beta$ | Residual Std. Dev. $\sigma_\varepsilon$|  
|-------|--------------------|------------------|----------------------------------------------|
| A     | 1.2%               | 1.1              | 7%                                           |
| B     | –0.4%              | 0.9              | 5%                                           |

The expected excess return on the market is $E(R_M$) = 8% and the standard deviation of the market’s excess returns is $\sigma_M$ = 18%.


a) What are the **expected excess returns** for Stocks A and B?  
b) What is the **covariance** between the returns of Stocks A and B?  
c) Consider an **equally weighted** portfolio of A and B. What are this portfolio’s:
   - $\alpha$,  
   - $\beta$,  
   - residual standard deviation, and  
   - total standard deviation?

---

## Solution 3

#### Given:

- Stock A: $\alpha_A = 0.012$,  $\beta_A = 1.1$, $\sigma_{\varepsilon,A} = 0.07$  
- Stock B: $\alpha_B = -0.004$, $\beta_B = 0.9$, $\sigma_{\varepsilon,B} = 0.05$ 
- Market: $E(R_M) = 0.08$, $\sigma_M = 0.18$

### a) Expected excess returns

For stock $i$:
$$
E(R_i) = \alpha_i + \beta_i E(R_M).
$$

- Stock A:
  $$
  E(R_A) = 0.012 + 1.1 \times 0.08
  = 0.012 + 0.088
  = 0.10
  = 10\%.
  $$

- Stock B:
  $$
  E(R_B) = -0.004 + 0.9 \times 0.08
  = -0.004 + 0.072
  = 0.068
  = 6.8\%.
  $$


### b) Covariance between A and B

Residuals are uncorrelated in the single-index model, so:
$$
\text{Cov}(A,B)
= \beta_A \beta_B \sigma_M^2
= 1.1 \times 0.9 \times 0.18^2
= 0.99 \times 0.0324
\approx 0.03208.
$$


### c) Equally weighted portfolio $(w_A = w_B = 0.5)$

- **Alpha**:
  $$
  \alpha_P = 0.5 \alpha_A + 0.5 \alpha_B
  = 0.5 \times 0.012 + 0.5 \times (-0.004)
  = 0.004
  = 0.4\%.
  $$

- **Beta**:
  $$
  \beta_P = 0.5 \beta_A + 0.5 \beta_B
  = 0.5 (1.1 + 0.9)
  = 1.0.
  $$

- **Residual variance**:
  $$
  \sigma_{\varepsilon,P}^2
  = 0.5^2 \sigma_{\varepsilon,A}^2
  + 0.5^2 \sigma_{\varepsilon,B}^2
  = 0.25 (0.07^2 + 0.05^2)
  = 0.25 (0.0049 + 0.0025)
  = 0.00185.
  $$

  Residual standard deviation:
  $$
  \sigma_{\varepsilon,P}
  = \sqrt{0.00185}
  \approx 0.0430
  = 4.30\%.
  $$

- **Total variance**:
  $$
  \sigma_P^2
  = \beta_P^2 \sigma_M^2 + \sigma_{\varepsilon,P}^2
  = 1^2 \times 0.18^2 + 0.00185
  = 0.0324 + 0.00185
  = 0.03425.
  $$

  Total standard deviation:
  $$
  \sigma_P
  = \sqrt{0.03425}
  \approx 0.1851
  = 18.51\%.
  $$

## Question 4 – Fixed Income
Ashley's company bond has a **face value of €1,000** and **6 years** remaining to maturity. The bond pays **annual coupons at 6%**. It is currently trading at a price that corresponds to a **current yield of 5.2%**.

a) Determine the bond’s **current market price**.

b) Compute the bond’s **Yield to Maturity**.

c) At the current price, calculate the **Macaulay Duration** and the **Modified Duration** of the bond.

d) Apply the **Modified Duration approximation** to estimate the percentage change in the bond’s price resulting from a 75-basis-point increase in the Yield to Maturity. Then, recalculate the bond’s **actual price change**. Compare the estimated and actual price changes, and explain why any deviation may or may not arise.

e) Calculate the bond’s **Convexity** at the current price.

f) Applying the **duration–convexity approximation**, estimate the percentage change in the bond’s price if the Yield to Maturity increases by 75 basis points.

---

## Solution 4

### Given
- Face value: $F = €1000$  
- Maturity: $n = 6$ years  
- Annual coupon rate: $c = 6%$
- Annual coupon:  
$$
C = cF = 0.06 \times 1000 = €60
$$
- Current yield:  
$$
CY = 5.2% = 0.052
$$


### a) Current market price (using current yield)

By definition,
$$
CY = \frac{C}{P}
$$

Solving for price:
$$
P = \frac{C}{CY} = \frac{60}{0.052} = 1153.846
$$

$$
\boxed{P \approx €1{,}153.85}
$$



### b) Yield to Maturity (YTM)

The Yield to Maturity $y$ solves:
$$
P = sum_{t=1}^{6} \frac{60}{(1+y)^t}
    + \frac{1000}{(1+y)^6}
$$

Using $P = 1153.846$, numerical solution gives:
$$
\boxed{y \approx 0.031463 = 3.146\% \text{ per year}}
$$


### c) Macaulay Duration and Modified Duration

**Macaulay Duration**
$$
D_{\text{Mac}} =
\frac{\sum_{t=1}^{6} t \cdot \frac{CF_t}{(1+y)^t}}{P}
$$

Substituting cash flows and \(y = 3.146\%\):
$$
\boxed{D_{\text{Mac}} \approx 5.2739 \text{ years}}
$$

**Modified Duration**
$$
D_{\text{mod}} = \frac{D_{\text{Mac}}}{1+y}
$$
$$
D_{\text{mod}} = \frac{5.2739}{1.031463} = 5.1130
$$

$$
\boxed{D_{\text{mod}} \approx 5.1130}
$$



### d) Duration-only estimate vs. actual price change ( +75 bps )

Yield change:
$$
\Delta y = +75 \text{ bps } = +0.0075
$$

#### (i) Modified Duration approximation

$$
\frac{\Delta P}{P} \approx - D_{\text{mod}} \, \Delta y
$$

$$
\frac{\Delta P}{P} \approx -(5.1130)(0.0075) = -0.03835
$$

$$
\boxed{\%\Delta P_{\text{Dur}} \approx -3.8347\%}
$$

Estimated euro price change:
$$
\Delta P_{\text{Dur}} \approx 1153.846 \times (-0.03835) = -44.25
$$

$$
\boxed{\Delta P_{\text{Dur}} \approx -€44.25}
$$


#### (ii) Actual price change (exact repricing)

New yield:
$$
y_1 = y + \Delta y = 0.031463 + 0.0075 = 0.038963
$$

Repriced bond value:
$$
P(y_1) =
\sum_{t=1}^{6} \frac{60}{(1+y_1)^t}
+ \frac{1000}{(1+y_1)^6}
$$

$$
P(y_1) \approx 1110.65
$$

Actual euro change:
$$
\Delta P_{\text{actual}} = 1110.65 - 1153.85 = -43.19
$$

$$
\boxed{\%\Delta P_{\text{actual}} \approx -3.7433\%}
$$


#### (iii) Comparison and explanation

- Duration-only estimate:  
  $$
  -3.8347\% \quad (\approx -€44.25)
  $$

- Actual repriced change:  
  $$
  -3.7433\% \quad (\approx -€43.19)
  $$

**Explanation:**  
Modified duration provides a **first-order (linear) approximation** to the bond’s **curved price–yield relationship**. Because standard coupon bonds exhibit **positive convexity**, the duration-only method **overestimates price losses** when yields increase. The deviation is modest here due to the moderate yield change and medium maturity.



### e) Convexity

Convexity (annual-pay definition):
$$
\text{Convexity} =
\frac{1}{P}
\sum_{t=1}^{6}
\frac{t(t+1) \, CF_t}{(1+y)^{t+2}}
$$

Using $y = 3.146\%$:
$$
\boxed{\text{Convexity} \approx 33.1391}
$$


### f) Duration–convexity approximation

$$
\frac{\Delta P}{P}
\approx
- D_{\text{mod}} \Delta y
+ \frac{1}{2} \text{Convexity} (\Delta y)^2
$$

$$
\frac{\Delta P}{P}
\approx
-(5.1130)(0.0075)
+ \frac{1}{2}(33.1391)(0.0075)^2
$$

$$
\frac{\Delta P}{P} \approx -0.037415
$$

$$
\boxed{\%\Delta P_{\text{Dur+Conv}} \approx -3.7415\%}
$$

Estimated price:
$$
P \approx 1153.85 \times (1 - 0.037415) = 1110.68
$$

This closely matches the exact repriced value, confirming that **convexity corrects the curvature error** in the duration-only estimate.

## Question 5 — Futures, Bonds, and Convenience Yield

a) A 1-year index futures trades at 4,200.  
Risk-free rate = 2%, dividend yield = 1.5% (discrete).  
Find the spot index level.

b)  Treasury bond:
- Spot price = 102
- Coupon = 3 paid in 6 months
- Risk-free rate = 6% (EAR)

Compute the 6-month futures price.

c) Commodity:
- Spot = 85
- 1-year futures = 82.5
- Risk-free rate = 3%

Estimate the convenience yield.

---


## Solution 5

#### (a) Spot index
$$
S_0 = F \frac{1+q}{1+r}
= 4200 \cdot \frac{1.015}{1.02}
\approx 4179.41
$$

#### (b) Bond futures
$$
F_0 = (102 - \frac{3}{1.06^{0.5}}) \cdot 1.06^{0.5}
\approx 102.02
$$

#### (c) Convenience yield
$$
y = r - \ln(F_0/S_0)
= 0.03 - \ln(82.5/85)
\approx 5.99\%
$$

## Question 6 - Two-Asset Portfolio (Mean–Variance)

You are analysing a portfolio composed of **two risky assets**, Asset A and Asset B.  
The characteristics of the assets are as follows:

- Expected return of Asset A: $E(r_A)=$ 14%
- Expected return of Asset B: $E(r_B)=$ 7%
- Standard deviation of Asset A: $ \sigma_A=$ 22%
- Standard deviation of Asset B: $ \sigma_B=$ 18%
- Correlation between the two assets: $ \rho_{AB}=-0.25$

Assume a risk-free rate of $r_f$ =2.5%.

a) Compute the **covariance** between Asset A and Asset B.

b) Starting from the **portfolio variance expression**, derive the **weight on Asset A** in the **global minimum-variance portfolio**.

c) Calculate the **minimum-variance portfolio weights** and compute the portfolio’s **expected return, and volatility**.

---

## Solution 6

### a) Minimum-Variance Portfolio (MVP): Derivation

Let $w_A$ be the weight on Asset A and $w_B=1-w_A$ .

Portfolio variance:
$$
\sigma_p^2
= w_A^2\sigma_A^2
+ (1-w_A)^2\sigma_B^2
+ 2w_A(1-w_A)\text{Cov}(A,B)
$$

Differentiate with respect to $w_A$ and set equal to zero:
$$
\frac{\partial \sigma_p^2}{\partial w_A}
= 2w_A\sigma_A^2
- 2(1-w_A)\sigma_B^2
+ 2(1-2w_A)\text{Cov}(A,B)
= 0
$$

Solving for $w_A$ gives the global MVP weight:
$$
w_A^{\text{MVP}}
= \frac{\sigma_B^2-\text{Cov}(A,B)}
{\sigma_A^2+\sigma_B^2-2\text{Cov}(A,B)}
$$


### b) Minimum-Variance Portfolio weights
The covariance is given by:
$$
\text{Cov}(A,B)=\rho_{AB}\sigma_A\sigma_B
$$

Substituting values:
$$
\text{Cov}(A,B)=(-0.25)(0.22)(0.18)=-0.0099
$$

$$
\boxed{\text{Cov}(A,B)=-0.0099}
$$

Compute variances:
$$
\sigma_A^2=0.22^2=0.0484,
\qquad
\sigma_B^2=0.18^2=0.0324
$$

Weight on Asset A:
$$
w_A
= \frac{0.0324-(-0.0099)}
{0.0484+0.0324-2(-0.0099)}
= \frac{0.0423}{0.1006}
= 0.4205
$$

Weight on Asset B:
$$
w_B=1-w_A=0.5795
$$


### Portfolio expected return
$$
E(r_p)
= 0.4205(0.14)+0.5795(0.07)
= 0.0994
$$

$
\boxed{E(r_p)=9.94\%}
$


### Portfolio variance and volatility
$$
\sigma_p^2
= (0.4205)^2(0.0484)
+ (0.5795)^2(0.0324)
+ 2(0.4205)(0.5795)(-0.0099)
$$

$$
\sigma_p^2=0.01459
\qquad\Rightarrow\qquad
\sigma_p=\sqrt{0.01459}=0.1208
$$

$
\boxed{\sigma_p=12.08\%}
$
