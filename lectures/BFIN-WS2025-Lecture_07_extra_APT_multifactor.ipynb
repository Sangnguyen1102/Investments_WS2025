## Lecture 07 - Extra Session: APT & Multifactor Models

---


### Goals for This Session

* Reinforce APT logic and its link to diversification.
* Connect theoretical APT to empirical factor models used in practice.
* Provide a slide-ready summary for teaching and attribution examples.

---


### Arbitrage Primer

* Arbitrage: zero-net-investment, riskless profit; competition removes it quickly.
* In well-functioning markets, persistent arbitrage opportunities should not exist.
* Sets up the pricing restriction that equal betas must earn equal expected excess returns.

### Diversification and Idiosyncratic Risk

* With many assets, idiosyncratic components cancel; residual risk approaches zero.
* Well-diversified portfolios are priced only by systematic factor exposure.
* APT no-arbitrage conditions apply strictly to these diversified portfolios.

---


### APT: Single-Factor Logic

* Two diversified portfolios with the same factor beta must have the same expected return.
* If returns differ, long the higher-return and short the lower-return: factor risk cancels, leaving riskless profit.
* Competitive trading forces prices until equal-betas earn equal expected excess returns.

### No-Arbitrage Pricing Ratio (Single Factor)

* Absence of arbitrage implies a constant price per unit of factor risk:
$$
\frac{E[R_A]}{\beta_A} = \frac{E[R_B]}{\beta_B} = \lambda
$$
* $(\lambda)$: factor risk premium; expected excess return for $(\beta=1)$.

### General APT Equation (Multifactor)

* For asset $(i)$ with factor loadings $(\beta_{ik}$):
$$
E[R_i] = \beta_{i1}\lambda_1 + \beta_{i2}\lambda_2 + \dots + \beta_{iK}\lambda_K
$$
* Linear in factor sensitivities; no need for a universal market portfolio.

---


### Factor-Mimicking Portfolios

* Estimate factor premia using traded, diversified tracking portfolios.
* Normalization: $(\beta_k = 1)$ to the target factor; Orthogonality: $(\beta_{j \neq k} \approx 0)$.
* Diversification: idiosyncratic risk minimized so the premium isolates $(\lambda_k)$.
* Expected excess return of the factor-mimicking portfolio equals the factor premium.

---


### CAPM vs. APT: Assumptions

* **CAPM:** Requires a mean-variance efficient market portfolio held by all investors; strong distribution/utility assumptions.
* **APT:** Requires only a factor structure, enough securities to diversify away idiosyncratic risk, and no persistent arbitrage; does not require the market portfolio to be observed or held by everyone.

### CAPM vs. APT: Applicability

* **CAPM:** Pricing relation applies to every individual asset; all must sit on the SML.
* **APT:** Pricing holds exactly for well-diversified portfolios; single-asset deviations may persist because of idiosyncratic risk that cannot be costlessly hedged.

---


### Implementation Problem and the Factor Zoo

* APT is silent on which factors are priced; factors must be identified empirically.
* Candidate factors should affect returns, earn non-zero premia, and be non-diversifiable.
* Empirical search has produced a large "factor zoo"; robustness and economic rationale matter.

### Measurement: Beta Drift and Adjusted Beta

* Betas are not stationary; high betas tend to mean-revert downward and low betas upward.
* Blume-style adjustment: \(\beta_{adj} = w \cdot \beta_{sample} + (1-w) \cdot 1\) with \(w\) often around 0.67.
* Reduces estimation error but does not fix missing-factor problems.

---



### Fama-French 3-Factor Model (FF3)

* Motivation: size and value anomalies left unexplained by CAPM.
* Time-series regression on excess returns:
$$
R_{i,t} = \alpha_i + \beta_{i,M} R_{M,t} + \beta_{i,SMB} SMB_t + \beta_{i,HML} HML_t + e_{i,t}
$$
* Pricing relation: $(E[R_i] = \beta_{i,M} \lambda_M + \beta_{i,SMB} \lambda_{SMB} + \beta_{i,HML} \lambda_{HML}); (\alpha=0)$ if the model holds.

### Factors Defined: MKT, SMB, HML

* **MKT:** Market excess return $(R_M - r_f)$.
* **SMB:** Small minus Big; long small-cap, short large-cap.
* **HML:** High minus Low; long value (high B/M), short growth (low B/M).
* Long-short construction hedges market exposure and isolates characteristic premia.

### Constructing Factors and Orthogonalizing

* Stylized long-short portfolios approximate pure factor exposure.
* Dirty beta issue: characteristics are correlated (e.g., size with volatility).
* Double sorting (e.g., 2x3 Size x Value) helps keep non-target betas near zero.

---




### Carhart 4-Factor Model (Momentum)

* Adds momentum factor WML (Winners minus Losers) to FF3.
* Regression:
$$
R_{i,t} = \alpha_i + \beta_{i,M} R_{M,t} + \beta_{i,SMB} SMB_t + \beta_{i,HML} HML_t + \beta_{i,WML} WML_t + e_{i,t}
$$
* Widely used for mutual fund performance attribution; tests whether alpha is just momentum exposure.

### Fama-French 5-Factor Model

* Adds RMW (profitability: Robust minus Weak) and CMA (investment: Conservative minus Aggressive).
* In many US samples, adding RMW and CMA weakens the standalone value factor (HML) once profitability and investment are controlled for.
* Extends coverage of expected return variation beyond size and value alone.

### Pastor-Stambaugh Liquidity Factor

* Liquidity risk: assets that fall most when market-wide liquidity dries up require a premium.
* Augments FF3 with a liquidity factor \(LIQ\):
$$
E(R_i) = \beta_M \lambda_M + \beta_{SMB} \lambda_{SMB} + \beta_{HML} \lambda_{HML} + \beta_{LIQ} \lambda_{LIQ}
$$
* High \(\beta_{LIQ}\): vulnerable in crises, higher expected return; low/negative \(\beta_{LIQ}\): flight-to-quality hedge, lower expected return.

---



### Practical Factor Attribution Workflow

* Select factors (MKT, SMB, HML, MOM, RMW, CMA, LIQ) relevant to the strategy.
* Collect excess returns and factor data; align by date and frequency.
* Run time-series regression with constant and chosen factors; inspect betas, R-squared, and alpha.
* Interpret: large R-squared implies returns are factor-driven; alpha near zero suggests no unexplained skill.

### FF5 Factor Premia and Explanations

* **Market (MKT):** Equity risk premium for bearing broad market downturns.
* **SMB (Size):** Premium for small-cap risk—distress, funding, liquidity constraints; can shrink in large-cap-heavy periods.
* **HML (Value):** Premium for value/distress exposure; weaker when profitability/investment are controlled (post-FF5 evidence).
* **RMW (Profitability):** Robust minus Weak; compensates for underperformance in downturns of low-profit firms.
* **CMA (Investment):** Conservative minus Aggressive; firms with aggressive investment are riskier (option-like), so conservative firms earn higher premia.
* **Interactions:** Premia vary over time/geography; macro conditions (recessions/liquidity) shift which premia dominate.

---



### Interpreting FF5 Coefficients

* **Beta MKT:** >0 means moves with the market; <0 is rare and would hedge market risk. Magnitude (>1 aggressive, <1 defensive) drives most variance.
* **Beta SMB:** >0 small-cap exposure; <0 large-cap exposure. Strongly negative often indicates quality/mega-cap bias.
* **Beta HML:** >0 value tilt; <0 growth/glamour tilt. Negative HML often pairs with higher duration/growth narratives.
* **Beta RMW:** >0 profitable-firm tilt; <0 weak-profit tilt (exposure to unprofitable names).
* **Beta CMA:** >0 conservative/low-investment tilt; <0 aggressive investment tilt (firms expanding rapidly).
* **Alpha (if included):** Return unexplained after the five factors. Statistically indistinguishable from 0 if the model captures the strategy.
* **R²:** Share of variation explained by factors. High R² ⇒ factor-driven; low R² ⇒ more idiosyncratic/unexplained.

### Question 1

**You are assessing the performance of the "Deep Value Explorer" Mutual Fund. Using monthly excess returns ($R$), you establish the following Fama-French 3-factor regression:**

$$ R_{Fund} = \alpha + 0.9(R_{Mkt}) + 0.8(R_{SMB}) + 0.6(R_{HML}) + e $$

**The average annualized factor premiums for the period are:**
*   Market ($R_M$) = 5.5%
*   Size ($R_{SMB}$) = 2.5%
*   Value ($R_{HML}$) = 3.5%
*   Risk-Free Rate = 4.0%

*Treat the factor premiums as annualized excess returns.*

-   Calculate the Fund's **Benchmarked Expected Return** (the total return required given its specific factor exposures).

-   The Fund reported an actual average annual total return of **11.5%**. What is the Fund's annualized Alpha? Did the manager add value?

-   Interpret the coefficients (+0.8 SMB and +0.6 HML). What specific segment of the market is this manager targeting?

---

### Question1:  Answer

*   **A. Benchmarked Expected Return:**
    *   First, calculate the required **Excess Return** using the factor exposures ($\beta$) and premiums ($\lambda$):
    *   $E[R_{Excess}] = \beta_M \lambda_M + \beta_{SMB} \lambda_{SMB} + \beta_{HML} \lambda_{HML}$
    
    *   $E[R_{Excess}] = 0.9(5.5\%) + 0.8(2.5\%) + 0.6(3.5\%)$
    
    *   $E[R_{Excess}] = 4.95\% + 2.0\% + 2.1\% = \mathbf{9.05\%}$
    
    *   **Total Required Return** = Risk-Free Rate + Required Excess Return
    
    *   $E[r_{Total}] = 4.0\% + 9.05\% = \mathbf{13.05\%}$

*   **B. Annualized Alpha:**
    *   $\alpha = \text{Actual Total Return} - \text{Required Total Return}$
    
    *   $\alpha = 11.5\% - 13.05\% = \mathbf{-1.55\%}$
    
    *   The manager did **not** add value. Despite earning a positive nominal return (11.5%), the manager underperformed the benchmark by 1.55% after adjusting for the risks taken. A passive portfolio with similar style exposures would have returned 13.05%.

*   **C. Style Interpretation:**
    *   **Positive SMB (+0.8):** The fund has a high positive sensitivity to the Small-Minus-Big factor. This indicates a heavy tilt toward **Small Cap** stocks.
    
    *   **Positive HML (+0.6):** The fund has a positive sensitivity to the High-Minus-Low factor. This indicates a tilt toward **Value** stocks (stocks with high book-to-market ratios).
    
    *   This manager is pursuing a **Small-Cap Value** strategy. The coefficients suggest the portfolio behaves similarly to small, undervalued companies.

### Question 2

You manage an aggressive tech portfolio ($P$) that is highly sensitive to Volatility risk ($\beta_{P, Vol} = 1.6$). You wish to neutralize this volatility exposure entirely while maintaining your positions.

You identify a **Defensive Strategy ETF (D)** that has the following factor sensitivities:
*   $\beta_{D, Mkt} = 0.6$

*   $\beta_{D, Vol} = -0.8$

-   Calculate the specific dollar amount of the Defensive ETF ($w_D$) you must add for every $1,000 invested in Portfolio P to eliminate the Volatility risk.

-   Your original Portfolio P has a Market Beta of 1.2. Calculate the **weighted average Market Beta** of the combined (hedged) portfolio.

---

### Question 2: Answer

*   **The Hedge Ratio ($w_D$):**
    *   To neutralize the exposure, the weighted beta of the combined position (Portfolio + Hedge) must equal zero for the specific factor.
    
    *   Formula: $w_D = - \frac{\beta_{P, Vol}}{\beta_{D, Vol}}$
    
        $$w_D = - \frac{1.6}{-0.8} = -(-2.0) = \mathbf{2.0}$$
    
    *   The hedge ratio is 2.0. This means for every \$1,000 in Portfolio P, you must add **\$2,000** worth of the Defensive ETF.

*   **The New Weighted Market Beta:**
    *   Beta is additive based on capital allocation.
    
    *   Total Capital = \$1,000 (Portfolio P) + \$2,000 (ETF D) = \$3,000.
    
    *   Weights:
        *   $W_P = \frac{1,000}{3,000} = \frac{1}{3}$
        *   $W_D = \frac{2,000}{3,000} = \frac{2}{3}$
    
        $$\beta_{Combined, Mkt} = W_P(\beta_{P, Mkt}) + W_D(\beta_{D, Mkt})$$
        $$\beta_{Combined, Mkt} = \frac{1}{3}(1.2) + \frac{2}{3}(0.6)$$
        $$\beta_{Combined, Mkt} = 0.4 + 0.4 = \mathbf{0.8}$$
    
    *   By adding the hedge, the overall Market Beta of your combined portfolio has dropped from 1.2 to 0.8, making it less sensitive to broad market movements while completely removing the Volatility factor risk.

### Question 3

Assume an APT framework with a single systematic risk factor. You are analyzing three well-diversified portfolios (Assume idiosyncratic risk is negligible).

| Portfolio | Beta ($\beta$) | Expected Return ($E[r]$) |
| :--- | :--- | :--- |
| **X** | 1.5 | 16% |
| **Y** | 0.5 | 8% |
| **Z** | 0.0 (Risk-Free) | 3% |

-   Determine if the market is in equilibrium. Specifically, verify if the "market price of risk" ($\lambda$) is consistent across portfolios X and Y relative to the risk-free rate.

-   If a discrepancy exists, identify which portfolio is relatively overvalued or undervalued. Construct a zero-investment, zero-beta arbitrage strategy to exploit this inefficiency using Portfolios X, Y, and Z. demonstrate the guaranteed profit.

---

### Question 3: Answer

*   **Identifying the Violation:**
    *   According to the APT, all well-diversified portfolios must satisfy the linear relationship: $E[r_i] = r_f + \beta_i \lambda$.
    
    *   Calculate the implied lambda ($\lambda$) for each risky portfolio:
        *   **Portfolio X:** $\lambda_X = \frac{E[r_X] - r_f}{\beta_X} = \frac{16\% - 3\%}{1.5} = \frac{13\%}{1.5} \approx \mathbf{8.67\%}$
        
        *   **Portfolio Y:** $\lambda_Y = \frac{E[r_Y] - r_f}{\beta_Y} = \frac{8\% - 3\%}{0.5} = \frac{5\%}{0.5} = \mathbf{10.0\%}$
    
    *   The market is not in equilibrium. Portfolio Y offers a higher reward-to-risk ratio (10%) than Portfolio X (8.67%). Therefore, Portfolio Y is undervalued (offers too much return for its risk), and Portfolio X is overvalued (offers too little return).

*   **The Arbitrage Strategy:**
    *   **Strategy:** Buy the "cheap" risk (Portfolio Y) and Sell the "expensive" risk (Portfolio X), neutralizing beta.
    
    *   **Beta Neutralization:**
        *   We want to Long Portfolio Y: Buy \$3 of Portfolio Y. (Risk exposure: $3 \times 0.5 = 1.5$ Beta).
        *   To hedge this beta, we must Short \$1 of Portfolio X. (Risk exposure: $-1 \times 1.5 = -1.5$ Beta).
        *   Net Beta: $1.5 - 1.5 = 0$.
    
    *   **Zero Investment:**
        *   Buy \$3 of Y (Cost: -\$3).
        *   Sell Short \$1 of X (Inflow: +\$1).
        *   Current Cash Balance: -\$2. Borrow \$2 at the risk-free rate (Asset Z) to fund the remaining purchase.
    
    *   **Payoff Calculation:**
        *   Long Y: $+3 \times 8\% = +0.24$
        *   Short X: $-1 \times 16\% = -0.16$
        *   Borrow Z: $-2 \times 3\% = -0.06$
        *   **Net Arbitrage Profit:** $0.24 - 0.16 - 0.06 = \mathbf{+0.02}$.
