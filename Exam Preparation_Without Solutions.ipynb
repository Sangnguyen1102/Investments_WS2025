b) Calculate the **fair value of a European Call Option** with a **strike price of $50** expiring in **2 periods**.

## Question 3 – Two Stocks with Single-Index Regressions

An analyst estimates the following single-index regressions of excess returns on the market’s excess returns:

| Stock | Alpha $\alpha$ | Beta $\beta$ | Residual Std. Dev. $\sigma_\varepsilon$|  
|-------|--------------------|------------------|----------------------------------------------|
| A     | 1.2%               | 1.1              | 7%                                           |
| B     | –0.4%              | 0.9              | 5%                                           |

The expected excess return on the market is $ E(R_M) = 8% $ and the standard deviation of the market’s excess returns is $\sigma_M = 18% $.


a) What are the **expected excess returns** for Stocks A and B?  
b) What is the **covariance** between the returns of Stocks A and B?  
c) Consider an **equally weighted** portfolio of A and B. What are this portfolio’s:
   - $\alpha$,  
   - $\beta$,  
   - residual standard deviation, and  
   - total standard deviation?

## Question 4 – Fixed Income
Ashley's company bond has a **face value of €1,000** and **6 years** remaining to maturity. The bond pays **annual coupons at 6%**. It is currently trading at a price that corresponds to a **current yield of 5.2%**.

a) Determine the bond’s **current market price**.

b) Compute the bond’s **Yield to Maturity**.

c) At the current price, calculate the **Macaulay Duration** and the **Modified Duration** of the bond.

d) Apply the **Modified Duration approximation** to estimate the percentage change in the bond’s price resulting from a 75-basis-point increase in the Yield to Maturity. Then, recalculate the bond’s **actual price change**. Compare the estimated and actual price changes, and explain why any deviation may or may not arise.

e) Calculate the bond’s **Convexity** at the current price.

f) Applying the **duration–convexity approximation**, estimate the percentage change in the bond’s price if the Yield to Maturity increases by 75 basis points.

a) Starting from the **portfolio variance expression**, derive the **weight on Asset A** in the **global minimum-variance portfolio**.

b) Calculate the **minimum-variance portfolio weights** and compute the portfolio’s **expected return, volatility**.
