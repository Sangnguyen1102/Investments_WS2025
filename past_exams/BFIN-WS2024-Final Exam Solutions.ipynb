### Question 1

**1. (24 points) A bond with a par value of €1,000 has a maturity of 5 years, makes 6% annual coupon payments, and has a current yield (not YTM) of 5.5%.**

**a)** What is the current price of the bond?

**b)** What is the YTM of the bond?

**c)** What is the duration of the bond at the current price?

**d)** What is the modified duration?

**e)** Estimate the price of the bond using the duration rule given a 25 basis points increase in the YTM.

**f)** What is the actual price of the bond when the YTM increases by 25 basis points?

**g)** What is the reason for the difference between the actual and estimated price of the bond? What would happen to the difference between the actual and estimated price change if it were a 20-year bond?

**h)** What is the convexity of the bond at its initial price?

**i)** Estimate the price of the bond using duration-convexity rule given a 25 basis points increase in the YTM.

---

### Question 1: Answer

**a) Current Price**
*   Formula: $\text{Current Yield} = \frac{\text{Annual Coupon}}{\text{Price}}$
*   Calculation:
    *   $P_B = \frac{60}{0.055} = €1,090.91$

**b) YTM**
*   Method: Solve the bond pricing equation for $y$.
*   Result: **3.9602%**

**c) Duration**
*   $D = 4.490724 \text{ years}$

**d) Modified Duration**
*   Formula: $D_{mod} = \frac{D}{1+y}$
*   Calculation:
    *   $D_{mod} = \frac{4.490724}{1.039602} = 4.319656$

**e) Duration Rule Estimate (+25 bps)**
*   Formula: $\% \Delta P \approx -D_{mod} \times \Delta y$
*   Calculation:
    *   $-4.319656 \times 0.0025 = -1.080\%$

**f) Actual Price Change**
*   New Price ($P_1$) at $y=4.2102\%$: €1,079.21
*   Actual $\% \Delta P$: $-1.07245\%$

**g) Reasoning**
*   **Convexity:** The price-yield relationship is convex, while duration is a linear approximation.
*   **Maturity Impact:** A 20-year bond exhibits higher convexity, which would result in a larger difference between the estimated and actual price change.

**h) Convexity**
*   Result: **24.001**

**i) Duration-Convexity Rule Estimate**
*   Formula: $\frac{\Delta P}{P} \approx -D_{mod}\Delta y + \frac{1}{2}[\text{Convexity} \times (\Delta y)^2]$
*   Calculation:
    *   $-1.080\% + \frac{1}{2}[24.001 \times (0.0025)^2] = -1.0725\%$
*   Estimated Price: $1,090.91 \times (1 - 0.010725) = €1,079.21$

### Question 2

**2. (20 points) A pension fund manager is considering three mutual funds. The first is a stock fund, the second is a long-term bond fund, and the third is a money market fund that provides a safe return of 8%. The characteristics of the risky funds are as follows:**

| | Expected Return | Standard Deviation |
| :--- | :--- | :--- |
| **Stock fund (S)** | 20% | 30% |
| **Bond fund (B)** | 12% | 15% |

**The correlation between the stock and bond fund returns is 0.10.**

**a)** If the weight of the stock fund in the optimal risky portfolio is 45.00%, what is the expected value, and standard deviation of the optimal risky portfolio’s rate of return?

**b)** What is the Sharpe Ratio of the optimal risky portfolio?

**c)** What would be your allocations to the risk-free asset and the optimal risky portfolio if your risk aversion coefficient is A=3?

**d)** What is the expected return and standard deviation of your complete portfolio?

**e)** You require that your complete portfolio yield an expected return of 18%, and to be efficient. What is the standard deviation of your portfolio? What are the weights of the money market fund and each of the two risky funds in your portfolio?

---

### Question 2: Answer

**a) Optimal Risky Portfolio Stats**
*   **Standard Deviation ($\sigma_p$):**
    *   $\sigma^2_p = w_S^2 \sigma_S^2 + w_B^2 \sigma_B^2 + 2 w_S w_B \text{Cov}(S, B)$
    *   $\sigma^2_p = 0.45^2(0.09) + 0.55^2(0.0225) + 2(0.45)(0.55)(0.10)(0.30)(0.15)$
    *   $\sigma^2_p = 0.02726 \rightarrow \sigma_p = 16.51\%$
*   **Expected Return ($E(r_p)$):**
    *   $E(r_p) = 0.45(20\%) + 0.55(12\%) = 15.6\%$

**b) Sharpe Ratio**
*   $S_p = \frac{15.6\% - 8\%}{16.51\%} = 0.4603$

**c) Allocation ($A=3$)**
*   $y^* = \frac{E(r_p) - r_f}{A \sigma_p^2} = \frac{0.156 - 0.08}{3(0.02726)} = 0.9293$
*   **Allocation:** 92.93% in the Optimal Risky Portfolio, 7.07% in the Risk-free Asset.

**d) Complete Portfolio**
*   $E(r_C) = 8\% + 0.9293(7.6\%) = 15.06\%$
*   $\sigma_C = 0.9293(16.51\%) = 15.34\%$

**e) Target Return (18%)**
*   Find leverage ($y$): $18\% = 8\% + y(7.6\%) \rightarrow y = 1.3158$
*   **Standard Deviation:** $1.3158 \times 16.51\% = 21.72\%$
*   **Weights:**
    *   Stock Fund: $1.3158 \times 0.45 = 59.21\%$
    *   Bond Fund: $1.3158 \times 0.55 = 72.37\%$
    *   Money Market: $1 - 1.3158 = -31.58\%$ (Borrowing)

### Question 3

**3. (16 points) Assume that the risk-free rate of interest is 6% and the expected rate of return on the market is 16%.**

**a)** A share of stock sells for \$50 today. It will pay a dividend of \$6 per share at the end of the year. Its beta is 1.2. What do investors expect the stock to sell for at the end of the year?

**b)** You are buying a firm with an expected perpetual cash flow of \$1,000 but are unsure of its risk. If you think the beta of the firm is .5, when in fact the beta is really 1, how much more will you offer for the firm than it is truly worth?

**c)** A stock has an expected rate of return of 4%. What is its beta?

**d)** The market price of a security is \$50. Its expected rate of return is 18%. What will 
be the market price of the security if its correlation coefficient with the market portfolio doubles (and all other variables remain unchanged)? Assume that the stock is expected to pay a constant dividend in perpetuity.

---

### Question 3: Answer

**a) Expected Stock Price**
*   Required Return (CAPM): $k = 6\% + 1.2(10\%) = 18\%$
*   Pricing: $0.18 = \frac{6 + P_1 - 50}{50} \rightarrow P_1 = \$53$

**b) Valuation Error**
*   Value at $\beta=0.5$ ($k=11\%$): $PV = 1000/0.11 = \$9,090.91$
*   Value at $\beta=1.0$ ($k=16\%$): $PV = 1000/0.16 = \$6,250$
*   **Difference:** You overestimate the value by offering \$2,840.91 more than it is worth.

**c) Beta Calculation**
*   $4\% = 6\% + \beta(10\%) \rightarrow \beta = -0.2$

**d) Correlation Impact**
*   Double Correlation $\rightarrow$ Double Beta ($\beta_{new} = 2.4$).
*   New Required Return: $k = 6\% + 2.4(10\%) = 30\%$.
*   Implied Dividend: $D = 50 \times 0.18 = \$9$.
*   New Price: $P = 9 / 0.30 = \$30$.

### Question 4

**4. (10 points) On your first day of trading in Vietnamese forward contracts, you observe that the share price of Giap Industries is currently VND 56,000 whereas the six-month forward price is VND 57,250. If the yield on a one-year riskless security is 5%, are arbitrage profits possible in this market? If not, explain why not. If so, devise an appropriate trading strategy.**

---


### Question 4: Answer

**Analysis**
*   Theoretical Forward Price:
    *   $F^* = S_0(1+r)^T = 56,000(1.05)^{0.5}$
    *   $F^* = \text{VND } 57,382.92$
*   **Conclusion:**
    *   $F^* > F_{market}$ (57,250). The market forward is undervalued.
    *   **Strategy:** Sell Spot, Lend proceeds, Buy Forward Contract.

### Question 5

**5. (10 points) The current stock price of BSA Corp is \$22 per share and the one-year risk-free interest rate is 6%. A one-year put on BSA Corp with a strike price of \$20 sells for \$4.25, while the identical call sells for \$7. Is there an arbitrage opportunity? If so, What trades would you do to exploit this arbitrage opportunity? What would be your arbitrage profit?**

---


### Question 5: Answer

*   Synthetic Call via Put-Call Parity:
    *   $C_{syn} = S_0 + P - PV(K)$
    *   $C_{syn} = 22 + 4.25 - \frac{20}{1.06} = \$7.382$

*   **Arbitrage:**
    *   $C_{syn} > C_{market}$ (\$7.00).
    *   **Strategy:** Buy Call, Short Synthetic Call (Short Stock, Short Put, Lend PV(K)).
    *   **Profit:** $\$0.382$

### Question 6

**6. (20 points) Consider a European put option that expires in one period and has an exercise price of \$50. Assume that the stock price today is equal to \$50. The one-period risk-free rate is 7.5%. In one period, the stock price will either rise by \$10 or fall by \$15.**

**a.** What are the risk-neutral probabilities?

**b.** What is the fair price of this option?

**c.** What is the replicating portfolio for this option?

**d.** What is the fair price of a call option with the same strike price?

---

### Question 6:Answer

**a) Risk-neutral probabilities**
*   $u = 1.2, d = 0.7$.
*   $p_u = \frac{1.075 - 0.7}{1.2 - 0.7} = 0.75$
*   $p_d = 0.25$

**b) Put Price**
*   Payoffs: $P_u=0, P_d=15$.
*   $f_0 = \frac{0.75(0) + 0.25(15)}{1.075} = \$3.488$

**c) Replicating Portfolio**
*   Hedge Ratio ($\Delta$): $\frac{0-15}{60-35} = -0.60$ (Short 0.6 shares).
*   Lending ($B$): $\$33.488$.

**d) Call Price**
*   $C = 50 + 3.488 - \frac{50}{1.075} = \$6.977$
