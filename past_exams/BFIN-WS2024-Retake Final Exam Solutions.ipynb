### Question 1

**1. (10 points) Suppose there are only two stocks in the world: stock A and Stock B. The expected returns of these two stocks are 10 percent and 20 percent, while the standard deviations of the stocks are 5 percent and 15 percent, respectively. The covariance between the two stocks is 0.0075.**

**a)** Suppose you want your portfolio to earn an average/expected return of 12%. What fraction of your investment should be in stocks A and B?

**b)** Would your portfolio in a) benefit from diversification? Make sure you provide the specifics of your reasoning.

---

### Question 1: Answer

**a) Weights**
*   $12\% = w_A(10\%) + (1-w_A)(20\%)$
*   $0.12 = 0.20 - 0.10w_A \rightarrow w_A = 0.80$
*   **Allocation:** 80% in Stock A, 20% in Stock B.

**b) Diversification**
*   Correlation $\rho_{AB} = \frac{0.0075}{0.05 \times 0.15} = 1.0$
*   **Conclusion:** No diversification benefit exists because the assets are perfectly positively correlated.

### Question 2

**2. (24 points) Elif and Eren are two individual investors with different risk preferences. Both of them hold diversified portfolios. Elif holds 40% of his wealth in T-bills and the rest in the market portfolio. Eren borrows 20% of his wealth and invests that amount, plus the rest of his wealth, in the market portfolio. Below is the forecasted distribution of market returns for the next year, as well as the distribution of ABC stock. The risk-free rate is 4% per year.**

| Probability | Market return | ABC |
| :--- | :--- | :--- |
| 0.2 | 6% | 4.0% |
| 0.3 | 8% | 2.5% |
| 0.3 | 12% | 3.0% |
| 0.2 | 14% | 2.0% |

**a)** What is the expected return of Elif’s and Eren’s portfolios?

**b)** Per unit of standard deviation risk, what risk-premium do Eren and Elif expect to receive on their portfolios?

**c)** ABC stock’s forecasted returns in this market are given above. Is ABC’s expected return in line with the expected return implied by CAPM? Why or why not? Show your work.

---

### Question 2: Answer

**a) Expected Returns**
*   $E(R_M) = 10\%$.
*   $E(R_{Elif}) = 0.4(4\%) + 0.6(10\%) = 7.6\%$
*   $E(R_{Eren}) = -0.2(4\%) + 1.2(10\%) = 11.2\%$

**b) Risk Premium per Unit of Risk (Sharpe Slope)**
*   Slope = $\frac{E(r_p) - r_f}{\sigma_p}$
*   Since both lie on the Capital Market Line (CML), the slope is identical.
*   Slope $\approx 2.0229$.

**c) CAPM Check (ABC Stock)**
*   $E(R)_{dist} = 2.85\%$
*   Calculated Beta $\approx -0.1477$
*   $E(R)_{CAPM} = 4\% - 0.1477(6\%) = 3.11\%$
*   **Conclusion:** ABC is not in equilibrium ($2.85\% \neq 3.11\%$).

### Question 3

**3. (8 points) A company enters into a short futures contract to sell 5,000 bushels of wheat for 450 cents per bushel. The initial margin is \$3,000 and the maintenance margin is \$2,000. What price change would lead to a margin call? Under what circumstances could \$1,500 be withdrawn from the margin account?**

---

### Question 3: Answer

**Margin Call (Short Position)**
*   **Position:** Short (Sell). Losses occur if the price **rises**.
*   **Threshold:** Equity < Maintenance Margin (\$2,000).
    *   Max Allowable Loss = Initial (\$3,000) - Maintenance (\$2,000) = \$1,000.
*   **Price Change Calculation:**
    *   Loss Formula: $(\text{Price}_{new} - \text{Entry}) \times \text{Quantity} = \text{Loss}$
    *   $(\Delta P_{cents} / 100) \times 5,000 = \$1,000$
    *   $\Delta P_{cents} = \frac{1000 \times 100}{5000} = 20 \text{ cents}$.
*   **Conclusion:** A margin call occurs if the price **rises** by 20 cents (Price $\ge$ 470 cents).

**Withdrawal**
*   **Condition:** Withdrawal allowed if Equity > Initial Margin + Withdrawal Amount.
*   Required Profit: \$1,500.
*   **Price Change Calculation:**
    *   Profit Formula: $(\text{Entry} - \text{Price}_{new}) \times \text{Quantity} = \text{Profit}$
    *   $(\Delta P_{cents} / 100) \times 5,000 = \$1,500$
    *   $\Delta P_{cents} = 30 \text{ cents}$.
*   **Conclusion:** \$1,500 can be withdrawn if the price **falls** by 30 cents (Price $\le$ 420 cents).

(8 points) On your first day of trading in Vietnamese forward contracts, you observe that the share price of Giap Industries is currently VND 58,000 whereas the six-month forward price is VND 59,750. If the yield on a one-year riskless security is 6%, are there any arbitrage opportunities in this market? If so, devise an appropriate trading strategy. If not, explain why not.**

---

### Question 4: Answer

*   Theoretical Forward Price:
    *   $F^* = 58,000(1.06)^{0.5} = 59,714.65$
*   **Conclusion:**
    *   $F^* < F_{market}$ (59,750). The market forward is overpriced.
    *   **Strategy:** Sell Forward (Market), Borrow PV(F), Buy Spot.
    *   **Profit:** VND 34.33.

### Question 5

**5. (10 points)**

**a. It is possible to buy six-month call options at \$5 and six-month puts at \$6 on stock Q. Both options have an exercise price of \$50. If the interest rate is 5% a year, what is the stock price?**

**b. A put option that expires in three months with an exercise price of \$72 sells for \$3.24. The stock is currently priced at \$70, and the risk-free rate is 4 percent per year, compounded continuously. What is the price of a call option with the same exercise price?**

---

### Question 5: Answer

**a) Implied Stock Price**
*   Formula (Parity): $S_0 = C - P + PV(K)$
*   Calculation: $S_0 = 5 - 6 + \frac{50}{(1.05)^{0.5}} = \$47.80$

**b) Call Price (Continuous)**
*   Formula: $C = S_0 + P - K e^{-rt}$
*   Calculation: $C = 70 + 3.24 - 72 e^{-0.04 \times 0.25} = \$1.96$

### Question 6

**6. (16 points) Consider a European put option that expires in one period and has an exercise price of \$52. Assume that the stock price today is equal to \$50. The one-period risk-free rate is 5%. In one period, the stock price will either rise or fall by 20%.**

**a.** What are the risk-neutral probabilities?

**b.** What is the fair price of this option?

**c.** What is the replicating portfolio for this option?

---

### Question 6: Answer

**a) Risk-neutral probabilities**
*   $u=1.2, d=0.8$.
*   $p_d = \frac{1.2 - 1.05}{1.2 - 0.8} = 0.375$
*   $p_u = 0.625$

**b) Fair Price**
*   Payoffs: $P_u=0$, $P_d=12$.
*   Price: $\frac{0.625(0) + 0.375(12)}{1.05} = \$4.29$

**c) Replicating Portfolio**
*   Hedge Ratio: $\frac{0-12}{60-40} = -0.60$ (Short 0.6 shares).
*   Lending: $\$34.29$.
