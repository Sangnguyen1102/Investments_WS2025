## 1. Empirical Failures of CAPM

CAPM pricing equation:

$$
E(R_i) = R_f + \beta_i \big(E(R_M) - R_f \big)
$$

where  
- $E(R_i)$ = expected return of asset $i$  
- $R_f$ = risk-free rate  
- $E(R_M)$ = expected market return  
- $\beta_i = \dfrac{\operatorname{Cov}(R_i, R_M)}{\operatorname{Var}(R_M)}$

The **alpha** of asset $i$ relative to CAPM is:

$$
\alpha_i = R_i - \Big[ R_f + \beta_i \big(R_M - R_f \big) \Big]
$$

Empirical evidence for **Systematic failures**:

- **Small-Firm Effect (SMB):** Small-cap stocks have **$\alpha_i > 0$** vs. CAPM.
- **Value Effect (HML):** High Book-to-Market (Value) stocks show **positive alphas**.
- Many portfolios earn returns **above** their CAPM-implied expected returns.

  
CAPM misprices assets because *market beta alone* cannot capture all relevant risk factors.

## 2. Intertemporal CAPM (ICAPM – Merton, 1973)
* Investors care about today's investment returns as well as how today's decisions affect future expected returns.
* ICAPM extends CAPM into a multi-period world.  
* The basic pricing relation for excess returns is:

$$
E(R_{i} - R_f) = \beta_{iM} \lambda_M + \sum_{k=1}^{K} \beta_{ik} \lambda_k
$$

where  
- $\beta_{iM}$ = sensitivity of asset $i$ to the market portfolio  
- $\beta_{ik}$ = sensitivity to state variable $k$ (e.g., interest rates, volatility, GDP growth)  
- $\lambda_M$ = market risk premium  
- $\lambda_k$ = risk premium for state variable $k$

**Key ideas:**

- Investors care about **current returns** and **future investment opportunities**.
- State variables include **interest rates, volatility, inflation, GDP growth**, etc.
- Assets that hedge against worsening opportunities (negative $\beta_{ik}$ to “bad” states) earn **lower risk premia**.

**Why CAPM fails here:**  
When multiple risks are present, CAPM (with only $\beta_{iM}$) produces pricing errors (alphas) because it ignores non-market factors.

## 3. Arbitrage Pricing Theory (APT)

***‘What must expected returns look like if there are no free arbitrage opportunities?’***

APT assumes that expected excess returns are **linear in factor loadings**:

$$
E(R_i) - R_f = \beta_{i1} \lambda_1 + \beta_{i2} \lambda_2 + \cdots + \beta_{iK} \lambda_K
$$

Realized excess returns can be written as:

$$
R_i - R_f = \beta_{i1} F_1 + \beta_{i2} F_2 + \cdots + \beta_{iK} F_K + \epsilon_i
$$

where  
- $F_k$ = realization of factor $k$ (mean $\lambda_k$)  
- $\beta_{ik}$ = sensitivity of asset $i$ to factor $k$  
- $\epsilon_i$ = idiosyncratic (diversifiable) error

If an asset is mispriced, we observe a **non-zero alpha**:

$$
\alpha_i = E(R_i) - \Big[ R_f + \sum_{k=1}^{K} \beta_{ik} \lambda_k \Big] \neq 0
$$

Arbitrageurs can construct **zero-net-investment portfolios** with:

- Zero factor exposure (cancel $\beta_{ik}$), but  
- Positive expected return (from $\alpha_i > 0$),

until arbitrage forces prices back to APT-consistent levels.

## 4. Fama–French 3-Factor Model

***Fama and French asked: "Which factors actually explain the anomalies we observe?"***
The FF3 model specifies excess returns as:

$$
R_i - R_f = \alpha_i + \beta_{iM} (R_M - R_f) + s_i \, \text{SMB} + h_i \, \text{HML} + \epsilon_i
$$

where  
- $R_i$ = return on asset/portfolio $i$  
- $R_M$ = market return  
- $\text{SMB}$ = Small Minus Big (size factor)  
- $\text{HML}$ = High Minus Low (value factor)  
- $\beta_{iM}$, $s_i$, $h_i$ = factor loadings  
- $\alpha_i$ = pricing error relative to FF3  
- $\epsilon_i$ = idiosyncratic error

**Interpretation:**

- CAPM: only $(R_M - R_f)$ explains excess returns.  
- FF3: adds **size** and **value** factors to capture small-firm and value premia.
- In many tests, $\alpha_i$ becomes statistically insignificant once SMB and HML are included.

## 5. Limits of Arbitrage & Behavioral Finance

Even if *risk-based* models are correct, prices can deviate from fundamentals because of:

### 5.1 Taste-Based Preferences (ESG Example)

Suppose investor utility depends on both **wealth** and a **“green” preference index** $G$:

$$
U = U(W_T, G)
$$

where  
- $W_T$ = terminal wealth  
- $G$ = share of green assets in the portfolio

An investor may accept a **lower expected return** for green portfolios:

$$
E(R_{\text{Green}}) < E(R_{\text{Brown}})
\quad \text{even if} \quad
\text{risk}(\text{Green}) \approx \text{risk}(\text{Brown})
$$

### 5.2 Efficiency Loss (Sharpe Ratio)

Sharpe ratio of a portfolio $p$:

$$
\text{SR}_p = \frac{E(R_p) - R_f}{\sigma_p}
$$

If investors exclude “Brown” assets, the **constrained green portfolio** often has:

$$
\text{SR}_{\text{Green-only}} < \text{SR}_{\text{Tangency}}
$$

showing an **efficiency cost** of non-financial preferences.


Behavioral biases and taste-based investing can sustain pricing anomalies even when arbitrage is conceptually possible.
